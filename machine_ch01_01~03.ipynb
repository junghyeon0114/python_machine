**행렬의 정렬**

np.sort( ) : 원행렬 유지한 채 정렬된 행렬 반환

ndarray.sort( ) : 원행렬을 정렬하여 반환

np.sort()[::-1] : 역순으로 정렬(내림차순)

image.png

```python
org_array=np.array([3,1,9,5])

print('org_array:', org_array)
print('np.sort 정렬된 형태:', np.sort(org_array))
print('np.sort 원본 형태:', org_array)
```

```python
#행열 정렬

array2=org_array.reshape(2,2)
print('array2:', array2)

#row 방향 정렬
print('row 방향:' , np.sort(array2, axis=0))
#col 방향 정렬
print('col 방향:', np.sort(array2, axis=1))
```
